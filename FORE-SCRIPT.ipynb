```python
import matplotlib .pyplot as plt
import pandas as pd
import numpy as np
```

```python
Data=pd.read_excel('WEATHER CALCULATOR.xlsx')
Data
```

```python
Data=pd.DataFrame(Data)
```

```python
len(Data)
```

```python
# creating a range for the data:

Data['HOURS']=pd.date_range(start='2022-10-27 12', end='2022-10-28 12', freq='3H')
Data=Data.set_index('HOURS')
# df['Notation']=pd.date_range(start='2021-10-15 12', end='2021-10-16 06', freq='3H')
```

```python
# Data_i=Data.drop[:,3:], axis=1
Data_2use=Data[['TEMP', 'RH']]
Data_2use
```

```python
fig,axes = plt.subplots(figsize=(10,5))
ax1=axes
ax1.plot(Data_2use.index, Data_2use.RH, color='red', marker="D")

ax2=axes.twinx()
ax2.plot(Data_2use.index,Data_2use.TEMP , color="blue", marker='D')

# axe.set_xlabel('', fontsize=20)
axes.set_xlabel('HOURS (UTC)', fontsize=20)
ax1.set_ylabel('TEMPERATURE ($^oC$)',color='red',  fontsize=20)
ax2.set_ylabel('RELATIVE HUMIDITY (%)',color='blue', fontsize=20)
axes.set_title('TEMPERATURE AND RELATIVE HUMIDITY-KUMASI',color='green', fontsize=12)
plt.savefig("Profile27.jpeg")
```

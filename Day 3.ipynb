```python
a=np.arange(0,361)
b=np.sin(np.deg2rad(a))
c= np.cos(np.deg2rad(a))
plt.plot(a,b, label='sin')
plt.plot(a,c, label='cos')

plt.legend()
plt.xlabel('angle')
plt.ylabel('trig values')
plt.title('Sin & Cos curves')
```

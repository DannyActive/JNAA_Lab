```python
# importing packages
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# reading files 
Accra = pd.read_fwf('Accra_1983_2017_dRR_gapless.txt',header= None, names=['Year', 'Month', 'Day', 'Precip'])
Kete_K = pd.read_fwf('Kete_Krachi_1983_2017_dRR_gapless.txt',header= None, names=['Year', 'Month', 'Day', 'Precip'])
Ksi = pd.read_fwf('Kumasi_1983_2017_dRR_gapless.txt',header= None, names=['Year', 'Month', 'Day', 'Precip'])
Wa = pd.read_fwf('Wa_1983_2017_dRR_gapless.txt', header=None, names=['Year', 'Month', 'Day', 'Precip'])
```

```python
# Accra
```

```python
# chamge to datetime

Accra['Date'] =pd.to_datetime(Accra[['Year', 'Month', 'Day']])
Kete_K['Date'] =pd.to_datetime(Kete_K[['Year', 'Month', 'Day']])
Ksi['Date'] =pd.to_datetime(Ksi[['Year','Month', 'Day']])
Wa['Date'] =pd.to_datetime(Wa[['Year','Month', 'Day']])
```

```python
Acc = Accra.drop(['Year', 'Month', 'Day'], axis=1).set_index('Date')
Kete = Kete_K.drop(['Year', 'Month', 'Day'], axis=1).set_index('Date')
Kuma = Ksi.drop(['Year', 'Month', 'Day'], axis=1).set_index('Date')
Wa_1 = Wa.drop(['Year', 'Month', 'Day'], axis=1).set_index('Date')
```

```python
Accr=Acc.resample('y').sum()
Ket=Kete.resample('y').sum()
Kumasi=Kuma.resample('y').sum()
Wa_2 = Wa_1.resample('y').sum()
```

```python
def ano (data):
    mean = np.mean(data)
    anomaly = data-mean
    return anomaly
```

```python
Ano_Accra=ano(Accr)
Ano_Kete = ano(Ket)
Ano_Ksi=ano(Kumasi)
Ano_Wa= ano(Wa_2)
```

```python
# assigning colours to the bar
Ano_Accra["color"] = np.where(Ano_Accra['Precip']<0, 'red', 'blue')

# plotting anomaly
fig = plt.bar(Ano_Accra.index, Ano_Accra['Precip'], width=200, color=Ano_Accra.color)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Year', fontsize=20)
plt.title('Rainfall Anomaly;Accra', fontsize=25)
```
